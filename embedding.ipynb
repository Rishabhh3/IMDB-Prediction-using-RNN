```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# Sentences
sent = ['the glass of milk',
    'the glass of juice',
    'The cup of tea',
    'I am good boy',
    'I am a good developer',
    'understand the meaning of words',
    'I am good',]
```

```python
# Define the vocabulary size

voc_size = 10000
```

```python
# one hot representation
one_hot_repr = [one_hot(word,voc_size)for word in sent]
one_hot_repr
```

The words are converted into these vectors
like in size of 10000 my 7087 index in 'The'
every word represnting one index
see ' The glass of' part is same in first and second

```python
# Cosine similarity is usually used in recommendation systems

# Word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# we want to make all sentence of same length to train,
# make max sentence length 

sent_length = 8
# add either 4 zeroes at forward or backward to make it of length 8
embedded_docs = pad_sequences(one_hot_repr, padding = 'pre', maxlen = sent_length)
```

```python
embedded_docs
```

```python
## Feature representation
dim= 10 # I want 10 features
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length = sent_length))
model.compile('adam','mse')
```

```python
model.predict(embedded_docs)
# every word will be represented by 10 dimenstions
```
